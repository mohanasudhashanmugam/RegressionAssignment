```python
import pandas as pd
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.svm import SVR
param_grid={'kernel':['rbf','linear'],'C':[0.01,0.1],'gamma':['auto','scale']}
grid=GridSearchCV(SVR(),param_grid,refit=True,verbose=3,n_jobs=-1,scoring='r2') 
```

```python


regressor=grid.fit(independent,dependent) 

```

```python
re=grid.cv_results_
print("The R_score value for best params {}",format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
y_pred=regressor.predict(independent)
```

```python
from sklearn.metrics import r2_score
r2_score=r2_score(dependent,y_pred)
r2_score
```
