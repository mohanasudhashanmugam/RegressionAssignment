```python
import pandas as pd
```

```python
dataset=pd.read_csv('insurance_pre.csv')
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.tree import DecisionTreeRegressor
param_grid={'criterion':['friedman_mse','mse','mae','poisson'],'splitter':['random','best'],'max_features':['auto','log2','sqrt',None]}
regressor=GridSearchCV(DecisionTreeRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1,scoring='r2') 

regressor=regressor.fit(independent,dependent)
```

```python
y_pred=regressor.predict(independent)
```

```python
from sklearn.metrics import r2_score
r2_score=r2_score(dependent,y_pred)
```

```python
r2_score
```

```python
import pickle
filename="Finalised_model_DT_grid.sav"
pickle.dump(regressor,open(filename,'wb'))
```
