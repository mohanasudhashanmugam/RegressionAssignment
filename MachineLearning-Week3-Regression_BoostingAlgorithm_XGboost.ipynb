```python
import pandas as pd
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30)
```

```python
##import xgboost as xgb
from xgboost import XGBRegressor

regressor=XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.1)

regressor.fit(x_train,y_train) 

```

```python
y_pred=regressor.predict(x_test)

from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
import pickle
filename="Finalised_model_XGBoost.sav"
pickle.dump(regressor,open(filename,'wb'))
```
